```python
from tqdm import tqdm
import requests
import pandas as pd
import numpy as np
```

```python
assays = [
    "ATG_PPARg_TRANS_dn", 
    "ATG_PPARg_TRANS_up", 
    "TOX21_PPARg_BLA_antagonist_ratio", 
    "TOX21_PPARg_BLA_Agonist_ratio", 
    "NVS_NR_hPPARg", 
    "ERF_ENZ_hCYP19A1_dn", 
    "NVS_ADME_hCYP19A1", 
    "NVS_ADME_hCYP19A1_Activator", 
    "TOX21_Aromatase_Inhibition", 
    "CEETOX_H295R_ESTRADIOL_dn", 
    "CEETOX_H295R_ESTRADIOL_noMTC_dn", 
    "CEETOX_H295R_ESTRADIOL_noMTC_up", 
    "CEETOX_H295R_ESTRADIOL_up", 
    "CEETOX_H295R_ESTRONE_dn", 
    "CEETOX_H295R_ESTRONE_noMTC_dn", 
    "CEETOX_H295R_ESTRONE_noMTC_up", 
    "CEETOX_H295R_ESTRONE_up"
]
```

```python
# find all matching

dfs = []

for assay in assays:
    path = f"data/{assay}.csv"
    df = pd.read_csv(path)
    
    df = df[["preferredName", "hitc"]] # only leave these two columns
    df.index = df["preferredName"] 
    del df["preferredName"] # set index to name and remove the column name
    df.columns = [assay] # set hitc (only remaining column) to name of assay
    
    dfs.append(df)
    
df = pd.concat(dfs, axis=1)
# TODO: no abs:
# df = df.apply(np.abs)
```

# Key event analysis

```python
key_events = [
    [
        "ATG_PPARg_TRANS_dn", 
        "ATG_PPARg_TRANS_up", 
        "TOX21_PPARg_BLA_antagonist_ratio", 
        "TOX21_PPARg_BLA_Agonist_ratio", 
        "NVS_NR_hPPARg"
    ],
    [
        "ERF_ENZ_hCYP19A1_dn", 
        "NVS_ADME_hCYP19A1", 
        "NVS_ADME_hCYP19A1_Activator", 
        "TOX21_Aromatase_Inhibition", 
    ], 
    [
        "CEETOX_H295R_ESTRADIOL_dn", 
        "CEETOX_H295R_ESTRADIOL_noMTC_dn", 
        "CEETOX_H295R_ESTRADIOL_noMTC_up", 
        "CEETOX_H295R_ESTRADIOL_up", 
        "CEETOX_H295R_ESTRONE_dn", 
        "CEETOX_H295R_ESTRONE_noMTC_dn", 
        "CEETOX_H295R_ESTRONE_noMTC_up", 
        "CEETOX_H295R_ESTRONE_up"
    ]
]

ke1 = df[key_events[0]]
ke2 = df[key_events[1]]
ke3 = df[key_events[2]]

events = pd.DataFrame({'MEI - PPARgamma': ke1.sum(axis=1),
                       'Aromatase': ke2.sum(axis=1),
                       'Estrogeni': ke3.sum(axis=1)})
```
